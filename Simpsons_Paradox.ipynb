# Simpson's Paradox

**Simpson's paradox** is a phenomenon that occurs when the results of a statistical analysis differ greatly when the affects of a confounding factor is not considered versus when it is. Hence, running statistical tests on the entire population would provide misleading results when compared to stratifying the population into smaller samples based on a confounding factor like Age, gender, location, etc. For example, analysis on the entire population may show that two variables are positively associated with each other, but the same analysis on sub-populations after stratification may show that these variables are negatively associated, or even independent of each other, hence the paradox. Considering that in real data, we often see a lot of confounding factors, not taking them into consideration would give contradictory and misleading results compared to the actual observation.

### Ethnicity Level Analysis

Here we will be analysing the Simpson's paradox in the California State DSS spending for the year 2014.
Upon aggregating the department's spending at an ethnicity level, we observe there exists a **huge disparity** between **_White not Hispanic_** and **_Hispanic_** communities.


### Analysis by Age Cohort

However, if we divide the dataset into smaller bins, by age, we come across a startling revelation.

It is observered that when stratified by age, there is infact **no disparity** between ethnicity (age is a confounding factor).

This is an example of Simpson's Paradox, wherein, a trend observed at the higher level disappears, or even reverses, when observed at a granular level.
